```python
import pandas as pd
from astropy.wcs import WCS
from astropy.io import fits

# Archivos de entrada y salida
input_csv = "coords.csv"   # Archivo de entrada con ID, RA, DEC
#fits_file = "../../../Downloads/belen_joined/NGC1313/NGC1313_image_i.fits"       # Imagen FITS con WCS
fits_file = "../"
output_csv = "coords_with_pixels.csv"  # Salida: CSV con X, Y
output_coo = "coords_iraf.coo"  # Salida: archivo .coo solo con RA y DEC

# Cargar el archivo CSV con coordenadas en grados
df = pd.read_csv(input_csv)

# Leer el header WCS de la imagen FITS
with fits.open(fits_file) as hdul:
    wcs = WCS(hdul[0].header)

# Convertir RA y DEC a coordenadas en píxeles (X, Y)
ra_dec_coords = zip(df['RA'], df['DEC'])
x, y = wcs.world_to_pixel_values(*zip(*ra_dec_coords))

# Agregar X e Y al DataFrame
df['X'] = x
df['Y'] = y

# Guardar el DataFrame completo en un CSV
df.to_csv(output_csv, index=False)
print(f"Archivo CSV con X e Y guardado en: {output_csv}")

# Crear archivo .coo con solo X y Y
df[['X', 'Y']].to_csv(output_coo, index=False, header=False, sep=' ')
print(f"Archivo .coo para IRAF guardado en: {output_coo}")
```

```python
import pandas as pd
import numpy as np

# Archivos de entrada y salida
input_csv = "ngc1313.csv"  # Archivo CSV de entrada
#output_csv = "filtered_coordinates.csv"  # Archivo CSV de salida

# Cargar el archivo CSV con coordenadas en grados
df = pd.read_csv(input_csv)

# RA de mi imagen: 49.73 a 49.4
# DEC de mi imagen: -66.56 a -66.43
ra_filter = df[(df['ra'] < 49.73) & (df['ra'] > 49.4)]
dec_filter = df[(df['dec'] > -66.56) & (df['dec'] < -66.43)]

# Filtrar filas que cumplan AMBOS criterios
filtered_df = df[((df['ra'] < 49.73) & (df['ra'] > 49.4)) & 
                 ((df['dec'] > -66.56) & (df['dec'] < -66.43))]

filtered_df
#Guardar las filas filtradas en un archivo CSV
#filtered_df.to_csv(output_csv, index=False)

#print(f"Tabla filtrada guardada en: {output_csv}")
```

```python
#quita los inf
filtered_df = filtered_df.loc[filtered_df['mag_i']!=np.inf]
```

```python
#quita los nan
filtered_df = filtered_df.loc[filtered_df["mag_i"].notna()]
```

```python
#busco el primer percentil (o percentil 10), de la distribución, so, las màs brillantes
np.percentile(filtered_df["mag_i"],10)
```

```python
#que me mande las menores al percentil 10 (es decir, màs brillantes porque menor magnitud aparente)
filtered_df = filtered_df.loc[filtered_df["mag_i"]<np.percentile(filtered_df["mag_i"],10)]
```

```python
#ascending false hace que me muestre de la menos brillante a la màs brillante (valores màs altos primero y al final màs bajos)
#luego yo eligo tail30 para elegir las últimas 30 que serían las últimas más brillantes (elijo mag 16 o 17 porque mag 20 son dèbiles para estrellas)
filtered_df = filtered_df.sort_values(by="mag_i",ascending=False).tail(30)
filtered_df
```

```python
output_csv = "filtered_coords_oficial.csv"  # Archivo CSV de salida
#UPDATE: este que tiene coords en vez de coors, este ahora tiene las 30 mas brillantes del percentil 10
#coors: este csv tiene: estrellas que sólo estén en mi imagen y además sólo toma las 80 menos brillantes del percentil 10 (más brillantes)
filtered_df.to_csv(output_csv, index=False)
```

Taylor+2011 stellar mass with photometry

$log M_*/[M_{sun}] = 1.15 + 0.70(g-i) - 0.4M_i$

```python
import numpy as np
import pandas as pd
data = pd.read_csv("masas_taylor11.csv")
g=data['mag_g']
i=data['mag_i']
M_i=data['Mag_i']
M_i
```

```python

#ahora, haremos el cálculo de Taylor 2011 para estimar la masa usando las bandas "g" e "i"
#ojo que él lo hace para galaxias para redshift intermedio (median z=0.2)
log_M_star_Msun = 1.15 + 0.70*(g-i) - 0.4*M_i
log_M_star_Msun
```

```python
log_M_star_Msun.mean()
```
