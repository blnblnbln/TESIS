```python
import numpy as np
import pandas as pd

# -----------------------------
# Parámetros de NGC 1313
# -----------------------------
RA0 = 49.566875       # grados
DEC0 = -66.498250     # grados
PHI = 0.0             # grados (PA)
INC = 48.0            # grados

# Distancias en Mpc
DISTANCES_MPC = [4.19, 6.51]

# -----------------------------
# Función Scarano+2008
# -----------------------------
def deprojected_radius_arcsec(ra_deg, dec_deg, ra0_deg, dec0_deg, phi_deg, inc_deg):
    """
    Calcula el radio deproyectado r en arcsec usando Scarano+2008
    """

    # Conversión a radianes
    ra = np.deg2rad(ra_deg)
    dec = np.deg2rad(dec_deg)
    ra0 = np.deg2rad(ra0_deg)
    dec0 = np.deg2rad(dec0_deg)

    phi = np.deg2rad(phi_deg)
    inc = np.deg2rad(inc_deg)

    # Diferencias angulares
    delta_ra = (ra - ra0) * np.cos(dec)
    delta_dec = dec - dec0

    # Pasar a arcsec
    delta_ra *= 206265.0
    delta_dec *= 206265.0

    # Fórmulas de Scarano+2008
    A = -(delta_ra * np.sin(phi)) + (delta_dec * np.cos(phi))
    B = -(delta_ra * np.cos(phi)) - (delta_dec * np.sin(phi))

    r_arcsec = np.sqrt(A**2 + (B**2) / (np.cos(inc)**2))

    return r_arcsec


def arcsec_to_kpc(r_arcsec, distance_mpc):
    """Convierte arcsec a kpc para una distancia dada"""
    distance_pc = distance_mpc * 1e6
    scale = distance_pc / 206265.0 / 1000.0  # kpc / arcsec
    return r_arcsec * scale


# -----------------------------
# Leer tabla CSV
# -----------------------------
# Se asume que el CSV tiene columnas: ID, RA, DEC
df = pd.read_csv("scarano_oficial.csv")

# -----------------------------
# Calcular r deproyectado
# -----------------------------
r_arcsec = deprojected_radius_arcsec(
    df["RA"].values,
    df["DEC"].values,
    RA0, DEC0,
    PHI, INC
)

df["r_arcsec"] = r_arcsec

# Distancias en kpc para cada distancia asumida
for d in DISTANCES_MPC:
    df[f"r_kpc_{d:.2f}Mpc"] = arcsec_to_kpc(r_arcsec, d)

# -----------------------------
# Guardar resultado
# -----------------------------
df.to_csv("regiones_HII_r_deproyectado.csv", index=False)

print("Archivo generado: regiones_HII_r_deproyectado.csv")
```
