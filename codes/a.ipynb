```python
from astropy.io import fits

# Abre el archivo FITS
file_path = '../../../reduccion/NGC1313/cal/ceNGC1313_MASK01_2D.fits'
hdul = fits.open(file_path)

# Imprime la información sobre las extensiones del archivo FITS
hdul.info()

# Extrae solo el slit 1 (SCI)
i = 2  # El índice de la primera extensión SCI

# Accede a la extensión de ciencia (SCI)
sci_data = hdul[i].data
sci_header = hdul[i].header

# Verifica si NAXIS2 es 1 y convierte los datos a 1D si es necesario
# Comprobamos si el array de datos es 2D y si la segunda dimensión (NAXIS2) es igual a 1
if sci_data.ndim == 2 and sci_data.shape[1] == 1:
    sci_data = sci_data.flatten()  # Convierte el array 2D a 1D
    sci_header['NAXIS'] = 1  # Establece NAXIS a 1
    sci_header['NAXIS1'] = sci_data.size  # Establece NAXIS1 al tamaño del array 1D
    if 'NAXIS2' in sci_header:
        del sci_header['NAXIS2']  # Elimina NAXIS2 del encabezado si existe

# Crea un PrimaryHDU con los datos SCI y su encabezado
primary_hdu = fits.PrimaryHDU(data=sci_data, header=sci_header)

# Guarda el nuevo archivo FITS
new_file_path = 'slit_1_SCI.fits'
primary_hdu.writeto(new_file_path, overwrite=True)
print(f'Archivo guardado: {new_file_path}')

# Cierra el archivo FITS original
hdul.close()
```

```python
from astropy.io import fits

# Abre el archivo FITS
file_path = '../../../reduccion/NGC1313/cal/ceNGC1313_MASK01_2D.fits'
hdul = fits.open(file_path)

# Imprime la información sobre las extensiones del archivo FITS
hdul.info()

# Extrae solo el slit 1 (SCI)
i = 2  # El índice de la primera extensión SCI

# Accede a la extensión de ciencia (SCI)
sci_data = hdul[i].data
sci_header = hdul[i].header

# Verifica si NAXIS2 es 1 y convierte los datos a 1D si es necesario
if sci_data.ndim == 2 and sci_data.shape[0] == 1 and sci_data.shape[1] != 1:
    sci_data = sci_data[0]  # Obtiene el primer (y único) elemento si NAXIS2 no es 1
    sci_header['NAXIS'] = 1  # Establece NAXIS a 1
    sci_header['NAXIS1'] = len(sci_data)  # Establece NAXIS1 al tamaño del array 1D
    if 'NAXIS2' in sci_header:
        del sci_header['NAXIS2']  # Elimina NAXIS2 del encabezado si existe

# Crea un PrimaryHDU con los datos SCI y su encabezado
primary_hdu = fits.PrimaryHDU(data=sci_data, header=sci_header)

# Guarda el nuevo archivo FITS
new_file_path = 'Slit_1_SCI.fits'
primary_hdu.writeto(new_file_path, overwrite=True)
print(f'Archivo guardado: {new_file_path}')

# Cierra el archivo FITS original
hdul.close()
```

```python
from astropy.io import fits

# Ruta del archivo FITS
file_path = '../../../reduccion/NGC1313/cal/ceNGC1313_MASK01_2D.fits'

# Abre el archivo FITS
with fits.open(file_path) as hdul:
    # Accede a la extensión de ciencia (SCI) del slit 1 (extensión 2)
    sci_data = hdul[2].data
    sci_header = hdul[2].header

    # Verifica si los datos tienen dos ejes y el segundo eje tiene tamaño 1
    if sci_data.ndim == 2 and sci_data.shape[1] == 1:
        # Aplana los datos a un solo eje
        sci_data = sci_data[:, 0]
        # Actualiza el encabezado
        sci_header['NAXIS'] = 1
        sci_header['NAXIS1'] = sci_data.size
        if 'NAXIS2' in sci_header:
            del sci_header['NAXIS2']

    # Crea un nuevo archivo FITS con una extensión PrimaryHDU
    fits.PrimaryHDU(data=sci_data, header=sci_header).writeto('slitt_1_SCI.fits', overwrite=True)
```
