```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
from astropy.wcs import WCS

hdul = fits.open('../../../reduccion/NGC1313/cal/ceNGC1313_MASK01_2D.fits')
data1 = hdul[2].data
h1 = hdul[2].header
obj_name1 = h1.get('OBJECT', 'Unknown')

flux1 = data1[0]
w1 = WCS(h1, naxis=1, relax=False, fix=False)
lam1 = w1.wcs_pix2world(np.arange(len(flux1)), 0)[0]

plt.plot(lam1, flux1)
plt.ylim(0, )
plt.title("Espectro slit 1")
plt.xlabel('Wavelength (Angstrom)')
plt.ylabel('Flux $(erg\ cm^{-2} s^{-1} A^{-1} · 1 x 10^{15})$')
```
