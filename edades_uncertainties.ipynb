```python
# ============================================
# Propagación de errores por Monte Carlo para Age(EW)
# Asumiendo: SB99['Solid'] = log10(EW [Å]) y SB99['Time'] = edad [años].
# Salida: CSV con ID, H_ALPHA_EQW, H_ALPHA_EQW_ERR, Age, Age_err (años).
# ============================================

import numpy as np
import pandas as pd
from scipy.interpolate import PchipInterpolator

# -----------------------------
# Parámetros
# -----------------------------
SB99_PATH = 'analysis_gmos/sb99a.dat'
EQW_PATH  = 'analysis_gmos/measurement_emission_lines_oficial.tsv'
CSV_OUT   = 'ages_mc_ngc1313.csv'

N_SAMPLES = 5000
RNG_SEED  = 42
ROWS_TO_DROP = [19, 20, 21, 22, 23]

# -----------------------------
# 1) SB99: Age(log10 EW) con PCHIP
# -----------------------------
sb = pd.read_csv(SB99_PATH, delim_whitespace=True, header=None)
sb.columns = ['Time', 'Solid', 'LongDashed', 'ShortDashed']

time_years = sb['Time'].to_numpy(float)
logEW_model = sb['Solid'].to_numpy(float)  # *** Confirmado: ya está en log10(EW [Å]) ***

# Verificación suave (no cambia las unidades; solo avisa si luce raro)
q05, q95 = np.nanquantile(logEW_model, [0.05, 0.95])
if not (-2.0 <= q05 and q95 <= 4.0):
    print("[Aviso] Los cuantiles de 'Solid' no parecen típicos de log10(EW). "
          "Verifique que 'Solid' sea log10(EW [Å]).")

# Limpiar y ordenar
m = np.isfinite(time_years) & np.isfinite(logEW_model)
x = logEW_model[m]
y = time_years[m]
order = np.argsort(x)
x = x[order]
y = y[order]

# PCHIP (sin duplicados en x)
x_unique, idx = np.unique(x, return_index=True)
y_unique = y[idx]
Age_of_logEW = PchipInterpolator(x_unique, y_unique, extrapolate=False)  # años

# -----------------------------
# 2) Mediciones y Monte Carlo
# -----------------------------
eqw = pd.read_csv(EQW_PATH, delimiter='\t')
eqw = eqw.drop(index=ROWS_TO_DROP, errors='ignore')

required = ['ID', 'H_ALPHA_EQW', 'H_ALPHA_EQW_ERR']
missing = [c for c in required if c not in eqw.columns]
if missing:
    raise ValueError(f"Faltan columnas requeridas: {missing}")

ID = eqw['ID'].to_numpy()
E  = eqw['H_ALPHA_EQW'].to_numpy(float)       # Å
sE = eqw['H_ALPHA_EQW_ERR'].to_numpy(float)   # Å

valid = np.isfinite(E) & np.isfinite(sE) & (E > 0.0) & (sE >= 0.0)
if not np.all(valid):
    print(f"[Aviso] {np.sum(~valid)} fila(s) con valores no físicos/NaN; se marcarán como NaN.")
E  = np.where(valid, E, np.nan)
sE = np.where(valid, sE, np.nan)

# Monte Carlo
rng = np.random.default_rng(RNG_SEED)
E_samp = rng.normal(E[:, None], sE[:, None], size=(len(E), N_SAMPLES))
E_samp = np.where(E_samp > 0.0, E_samp, np.nan)  # truncación a valores físicos
logE_samp = np.log10(E_samp)

Age_samp = Age_of_logEW(logE_samp)  # años

# Percentiles (16–50–84) y error simétrico equivalente 1σ
Age_p16 = np.nanpercentile(Age_samp, 16, axis=1)
Age_p50 = np.nanpercentile(Age_samp, 50, axis=1)  # estimador recomendado
Age_p84 = np.nanpercentile(Age_samp, 84, axis=1)
Age_err = 0.5 * (Age_p84 - Age_p16)

# Calidad (bordes del modelo o σEW grande)
frac_valid = np.sum(np.isfinite(Age_samp), axis=1) / float(N_SAMPLES)
if np.any(frac_valid < 0.80):
    print("[Nota] Algunas filas tienen <80% de muestras válidas (posible borde del modelo/σ grande). "
          "Considere reportar límites unilaterales para esas fuentes.")

# -----------------------------
# 3) Salida
# -----------------------------
out = pd.DataFrame({
    'ID': ID,
    'H_ALPHA_EQW': eqw['H_ALPHA_EQW'],
    'H_ALPHA_EQW_ERR': eqw['H_ALPHA_EQW_ERR'],
    'Age': Age_p50,    # años
    'Age_err': Age_err # años (1σ equivalente simétrico)
})
out.to_csv(CSV_OUT, index=False)
print(f"[OK] CSV escrito en: {CSV_OUT}")
```

```python
import numpy as np
import pandas as pd
from scipy.interpolate import PchipInterpolator

# =========================
# Configuración
# =========================
SB99_FILE = "analysis_gmos/sb99a.dat"  # 1a col: edad (yr), 2a col: log10(EW_Halpha)
EQW_FILE  = "analysis_gmos/measurement_emission_lines_oficial.tsv"  # columnas: ID, H_ALPHA_EQW, H_ALPHA_EQW_ERR
OUT_FILE  = "ages_from_EW_intervals.tsv"
K_SIGMA   = 1.0  # 1σ; para ~95% usa 1.96

# (opcional) replicar tu filtrado previo por índices si aplica
INDICES_A_REMOVER = [19, 20, 21, 22, 23]

# =========================
# 1) Cargar SB99 y construir f_age(log10EW) -> edad(yr)
# =========================
sb = pd.read_csv(SB99_FILE, delim_whitespace=True, header=None, comment="#")
if sb.shape[1] < 2:
    raise ValueError("sb99a.dat debe tener al menos dos columnas: edad(yr), log10(EW).")

sb = sb.iloc[:, [0, 1]].copy()
sb.columns = ["age_yr", "log10_EW_model"]

# Ordenar por x y colapsar duplicados (PCHIP requiere x estrictamente creciente)
sb = sb.sort_values("log10_EW_model").groupby("log10_EW_model", as_index=False)["age_yr"].mean()

x = sb["log10_EW_model"].values  # dominio: log10(EW)
y = sb["age_yr"].values          # codominio: edad en años

f_age = PchipInterpolator(x, y, extrapolate=False)

# =========================
# 2) Cargar tus mediciones
# =========================
eqw = pd.read_csv(EQW_FILE, sep="\t")
to_drop = [i for i in INDICES_A_REMOVER if i in eqw.index]
if to_drop:
    eqw = eqw.drop(index=to_drop)

EW  = eqw["H_ALPHA_EQW"].astype(float).values
dEW = eqw["H_ALPHA_EQW_ERR"].astype(float).values

# =========================
# 3) Calcular edades: central y bordes (EW ± Kσ)
# =========================
tiny = np.finfo(float).tiny
EW_low  = np.maximum(EW - K_SIGMA*dEW, tiny)   # -> más viejo
EW_high = EW + K_SIGMA*dEW                     # -> más joven

logEW      = np.log10(EW)
logEW_low  = np.log10(EW_low)
logEW_high = np.log10(EW_high)

age_c   = f_age(logEW)        # central
age_lo  = f_age(logEW_high)   # límite joven (EW+)
age_hi  = f_age(logEW_low)    # límite viejo  (EW-)

# =========================
# 4) Empaquetar y guardar
# =========================
out = eqw[["ID", "H_ALPHA_EQW", "H_ALPHA_EQW_ERR"]].copy()
out["Age_yr"]     = age_c
out["Age_yr_lo"]  = age_lo
out["Age_yr_hi"]  = age_hi
out["Age_Myr"]    = out["Age_yr"] / 1e6
out["Age_Myr_lo"] = out["Age_yr_lo"] / 1e6
out["Age_Myr_hi"] = out["Age_yr_hi"] / 1e6

# Errores asimétricos (en Myr)
out["Age_err_low_Myr"]  = out["Age_Myr"] - out["Age_Myr_lo"]  # debe ser >0
out["Age_err_high_Myr"] = out["Age_Myr_hi"] - out["Age_Myr"]  # debe ser >0

# Flags por si algún punto cae fuera del rango de la malla (PCHIP -> NaN)
out["flag_out_of_range"] = (
    out[["Age_yr", "Age_yr_lo", "Age_yr_hi"]].isna().any(axis=1)
)

out.to_csv(OUT_FILE, sep="\t", index=False)
print(f"Guardado: {OUT_FILE}")

# =========================
# 5) Ejemplo: 1962 ± 47 Å
# =========================
EW_ex, dEW_ex = 1962.0, 47.0
age_ex      = f_age(np.log10(EW_ex))
age_ex_lo   = f_age(np.log10(EW_ex + K_SIGMA*dEW_ex))
age_ex_hi   = f_age(np.log10(max(EW_ex - K_SIGMA*dEW_ex, tiny)))

print("\nEjemplo 1962±47 Å:")
print(f"  Edad central: {age_ex/1e6:.3f} Myr")
print(f"  EW+{K_SIGMA}σ -> {age_ex_lo/1e6:.3f} Myr (más joven)")
print(f"  EW-{K_SIGMA}σ -> {age_ex_hi/1e6:.3f} Myr (más vieja)")
print(f"  Intervalo (~{K_SIGMA}σ): -{(age_ex-age_ex_lo)/1e6:.3f} / +{(age_ex_hi-age_ex)/1e6:.3f} Myr")
```
